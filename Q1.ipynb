```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
link = requests.get('https://en.wikipedia.org/wiki/Main_Page')
```

```python
link.text
```

```python
bs = BeautifulSoup(link.text,"html.parser")
```

```python
header = bs.find_all(['h1','h2','h3','h4','h5','h6'])
```

```python
print("List of header tags : ", *header, sep='\n' )
```
